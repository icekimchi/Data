```python
branches = ['CSE', 'Mech', 'Electronics']

fy = [23,17,35]
sy = [20, 23, 30]
ty = [30,20,15]

trace1 = go.Bar(
   x = branches,
   y = fy,
   name = 'FY'
)
trace2 = go.Bar(
   x = branches,
   y = sy,
   name = 'SY'
)
trace3 = go.Bar(
   x = branches,
   y = ty,
   name = 'TY'
)

data = [trace1, trace2, trace3]
layout = go.Layout(barmode = 'group', title='Departments')
fig = go.Figure(data = data, layout = layout)
fig.show()
```

```python
Heart_Rate = [70, 35]
QOL = [59, 40]
Mortality = [9, 20]

branches = ['Group1', 'Group2']
```

```python
trace1 = go.Bar(
    y = branches,
    x = Heart_Rate,
    name = 'Resting Heart Rate 110 bpm',
    orientation = "h",
    text = Heart_Rate,
    textposition = 'auto',
)

trace2 = go.Bar(
    y = branches,
    x = QOL,
    name = 'Improved QOL',
    orientation = "h",
    text = QOL,
    textposition = 'auto',

)

trace3 = go.Bar(
    y = branches,
    x = Mortality,
    name = 'Mortality',
    orientation = "h",
    text = Mortality,
    textposition = 'auto',
)
```

```python
data = [trace1, trace2, trace3]
layout = go.Layout(barmode = 'group', title='Comparison Between Groups')
fig = go.Figure(data = data, layout = layout)

fig.update_layout(
            legend_x = 0,
            legend_y = 0,
            legend_orientation = "h"
            )

fig.show()
```

# Scatter Plot

```python
import plotly.graph_objs as go
import numpy as np

N = 100

x_vals = np.linspace(0, 1, N)

y1 = np.random.randn(N) + 5
y2 = np.random.randn(N)
y3 = np.random.randn(N) - 5
```

```python
trace0 = go.Scatter(
   x = x_vals,
   y = y1,
   mode = 'markers', #점으로 나타내기
   name = 'markers'
)
trace1 = go.Scatter(
   x = x_vals,
   y = y2,
   mode = 'lines+markers', #선에 점으로 표시
   name = 'line+markers'
)
trace2 = go.Scatter(
   x = x_vals,
   y = y3,
   mode = 'lines',
   name = 'line'
)
```

```python
data = [trace0, trace1, trace2]
fig = go.Figure(data = data)
fig.show()
```

# Plotly Express

```python
import plotly.express as px

df = px.data.tips()

df.head()
```

```python
df.tail()
```

```python
fig = px.box(df, y="total_bill")
fig.show()
```

```python
fig = px.box(df, x="time", y="total_bill")
fig.show()
```

```python
fig = px.box(df, x="time", y="total_bill", color="smoker",
             notched=True, # used notched shape 더 몰려있음을 나타냄
             title="Box plot of total bill",
             hover_data=["day"] # add day column to hover data
            )
fig.show()
```

```python
fig = px.violin(df, y="total_bill")
fig.show()
```

```python
fig = px.violin(df, y="tip", 
                x="smoker", 
                color="sex", 
                box=True, 
                points="all",
                hover_data=df.columns)
fig.show()
```

```python
fig = px.scatter(df, x="total_bill", 
                    y="tip", 
                    color="smoker", 
                    facet_col="sex") #facet_col : 몇 개를 만들건지? sex가 두 개이기 때문에, 두 개의 그래프가 나타난다.

fig.show()
```

```python
fig = px.scatter(df, x="total_bill", 
                    y="tip", 
                    color="smoker", 
                    facet_col="sex",
                    trendline="ols") #facet_col : 몇 개를 만들건지? sex가 두 개이기 때문에, 두 개의 그래프가 나타난다.

fig.show()
```

```python
fig = px.histogram(df, x="total_bill")
fig.show()
```

```python
fig = px.histogram(df, x="total_bill", color="sex")
fig.show()
```

```python
fig = px.histogram(df, x="total_bill", 
                    y="tip", color="sex",
                    marginal="box", # or violin, rug
                    hover_data=df.columns)

fig.show()
```

```python
fig = px.density_heatmap(df, x="total_bill", y="tip")
fig.show()
```

```python
fig = px.density_heatmap(df, x="total_bill", 
                         y="tip", 
                         marginal_x="histogram",
                         marginal_y="histogram")
fig.show()
```

```python
#모듈 import
import plotly.graph_objects as go
from plotly.subplots import make_subplots


media = ["Facebook", "Twitter", "LinkeIn", "Google+", "Pinterest"]
year_2011 = [59, 21, 10, 5, 5]
year_2012 = [49, 23, 12, 10, 6]
year_2013 = [38, 25, 15, 14, 8]
```

```python
fig = make_subplots(rows=1, cols=3,
                    specs=[[{'type':'domain'}, {'type':'domain'}, {'type':'domain'}]],
                    subplot_titles=("2011", "2012", "2013"))
```

```python
fig.add_trace(go.Pie(
    labels=media, 
    values=year_2011,
    name="2011"),
              row=1, col=1)

fig.add_trace(go.Pie(
    labels=media, 
    values=year_2012,
    name="2012"),
              row = 1, col = 2)

fig.add_trace(go.Pie(
    labels=media, 
    values=year_2013,
    name="2013"),
              row = 1, col = 3)

fig.update_traces(hoverinfo="label+percent+name")

fig.update_layout(title_text="Favorite Social Media Channel")

fig.show()
```
